## Image Segmentation

```python
segmented = cv2.imread('dog-seg.png')
segmented = cv2.cvtColor(segmented,cv2.COLOR_BGR2RGB)
plt.imshow(segmented)
plt.show()

edges = cv2.Canny(segmented,100,200)

plt.imshow(edges,cmap='gray')
plt.show()
```

```python
edges_inv = (255 - edges)
```

```python
plt.imshow(edges_inv,cmap='gray')
plt.show()
```

```python
dist_transform = cv2.distanceTransform(edges_inv,cv2.DIST_L2,0) 
plt.imshow(dist_transform,cmap='gray')
```

```python
Image  = segmented
```

```python
SIZE_X = Image.shape[1]
SIZE_Y = Image.shape[0]
SIZE = min(SIZE_X,SIZE_Y)


LINEWIDTH = 0.1

PI = 3.14159265359
TWOPI = 2.0*PI
BACK = [0,0,0,1]

RES = 'out1.png'
```

## Functions for creating circles

```python
import time
```

```python

is_fill = np.zeros([SIZE_X+1,SIZE_Y+1,2])

def is_ok(x1, y1, r1, x2, y2, r2):
    return ((x1-x2)**2 + (y1-y2)**2) > (r1 + r2)**2

def add_new_circles(circles,r,t):
    for x in range(2*r,SIZE_X- r):
        for y in range(2*r,SIZE_Y - r):
            a=True
            xr=x
            yr=y
            rr=r
            if dist_transform[y,x]>t:
                for c in circles:
                    if not is_ok(xr,yr,rr,c['x'],c['y'],c['r']):
                        a=False
                        break
                    
                if a==True:
                    circles.append({
                    'x': x, 
                    'y': y, 
                    'r': r,
                    })                   

      

    

def show(Image,ctx, circles):
    for c in circles:
        rgb = Image[int(c['y']),int(c['x'])]
        rgba = [rgb[0]/255, rgb[1]/255,rgb[2]/255, 1.0]
        ctx.set_source_rgba(*rgba)
        ctx.arc(c['x'], c['y'], c['r'], 0, TWOPI)
        ctx.fill()
        ctx.set_source_rgba(*[0,0,0,1])
        ctx.arc(c['x'], c['y'], c['r'], 0, TWOPI)
        ctx.stroke()
```
